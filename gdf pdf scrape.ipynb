```python
#cleaning up
price_clean = price2.dropna()
price_clean
```

```python
# Function to convert range to average
def range_to_avg(price_str):
    if isinstance(price_str, str) and '-' in price_str:
        start, end = price_str.split('-')
        return (float(start) + float(end)) / 2
    else:
        return float(price_str)

# Apply the function to the 'price' column
price_clean['price'] = price_clean['price'].apply(range_to_avg)

price_clean
```

```python
price_final = price_clean.drop(columns=['price_A', 'price_B'])
price_final
```
